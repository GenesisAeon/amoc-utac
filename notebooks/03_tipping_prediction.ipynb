# Notebook 03 — AMOC Tipping-Year Prediction

Compare UTAC-based tipping estimates with published literature:

| Source | Central year | Range |
|--------|-------------|-------|
| Ditlevsen & Ditlevsen (2023) | **2057** | 2025–2095 |
| Chavent et al. (2026) | ≤ 2100 | 50% weakening |
| **UTAC Package 18** | computed | 5–95% MC ensemble |

```python
import matplotlib.pyplot as plt
import numpy as np
from amoc_utac.tipping_predictor import TippingPredictor

pred = TippingPredictor(seed=42)
result = pred.predict_tipping_year(start_year=2024)

print('UTAC Tipping-Year Prediction')
print(f"  Central year   : {result['utac_central_year']:.0f}")
print(f"  5–95% range    : {result['utac_5pct']:.0f} – {result['utac_95pct']:.0f}")
print(f"  Ditlevsen 2023 : {result['ditlevsen_2023_central']}  "
      f"({result['ditlevsen_2023_range'][0]}–{result['ditlevsen_2023_range'][1]})")
print(f"  Chavent 2026   : ≤{result['chavent_2026_50pct_by']} (50% weakening)")
```

```python
# UTAC ODE phase portrait
gamma_values = np.linspace(0.0, 0.6, 300)
K = 18.0
sigma = 2.2
h_star = K * np.tanh(sigma * gamma_values)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(gamma_values, h_star, lw=2.5, color='steelblue', label='H* = K·tanh(σΓ)')
ax.axhline(9.0, color='tomato', ls='--', lw=1.5, label='50% threshold (9 Sv)')
ax.axvline(0.251, color='darkorange', ls=':', lw=1.5, label='Γ_AMOC = 0.251')
ax.scatter([0.251], [K * np.tanh(sigma * 0.251)],
           color='darkorange', zorder=5, s=80, label=f'Present setpoint')
ax.set_xlabel('CREP Γ')
ax.set_ylabel('Fixed point H* [Sv]')
ax.set_title('UTAC Fixed-Point Diagram: AMOC Tipping Landscape')
ax.legend()
plt.tight_layout()
plt.savefig('tipping_prediction.png', dpi=150)
plt.show()
```
