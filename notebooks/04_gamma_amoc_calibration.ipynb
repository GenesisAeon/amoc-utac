$$\Gamma_{\text{AMOC}} = \frac{\text{arctanh}(\eta_{\text{AMOC}})}{\sigma} = \frac{\text{arctanh}(0.50)}{2.2} \approx 0.251$$

```python
import math
import matplotlib.pyplot as plt
import numpy as np

from amoc_utac.constants import GAMMA_AMOC, UTAC_SIGMA

# Verify formula
gamma_computed = math.atanh(0.50) / UTAC_SIGMA
print(f'arctanh(0.50) / 2.2 = {gamma_computed:.6f}')
print(f'Stored GAMMA_AMOC   = {GAMMA_AMOC:.6f}')
print(f'Match: {abs(gamma_computed - GAMMA_AMOC) < 1e-6}')
```

```python
# CREP Spectrum Atlas visualisation
spectrum = {
    'Solar flares\n(P21)':      0.014,
    'Cygnus X-1\n(P17)':        0.046,
    'Amazon\n(P19)':            0.116,
    'AMOC\n(P18) ★':            0.251,
    'Neural\n(P20)':            0.251,
    'BTW\n(P22)':               0.296,
    'Manna\n(P22)':             0.376,
    'ERA5 Arctic\n(P01)':       0.920,
}

labels = list(spectrum.keys())
gammas = list(spectrum.values())
colours = ['gold' if '★' in l else 'steelblue' for l in labels]

fig, ax = plt.subplots(figsize=(12, 4))
bars = ax.barh(labels, gammas, color=colours, edgecolor='black', linewidth=0.6)
ax.axvline(0.251, color='darkorange', ls='--', lw=2, label='Γ = 0.251 (universal setpoint)')
ax.set_xlabel('CREP Γ', fontsize=12)
ax.set_title('GenesisAeon CREP Criticality Spectrum — Packages 17–22', fontsize=13)
ax.legend()
plt.tight_layout()
plt.savefig('crep_spectrum_atlas.png', dpi=150)
plt.show()
```

```python
# η → Γ inversion curve
eta = np.linspace(0.01, 0.99, 500)
gamma_curve = np.arctanh(eta) / UTAC_SIGMA

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(eta, gamma_curve, lw=2.5, color='steelblue', label='Γ = arctanh(η) / σ')
ax.scatter([0.50], [GAMMA_AMOC], color='tomato', zorder=5, s=120,
           label=f'AMOC: η=0.50 → Γ={GAMMA_AMOC:.3f}')
ax.set_xlabel('Efficiency η = H*/K', fontsize=11)
ax.set_ylabel('CREP Γ', fontsize=11)
ax.set_title('UTAC Efficiency → CREP Mapping (σ = 2.2)', fontsize=12)
ax.legend()
plt.tight_layout()
plt.savefig('eta_gamma_mapping.png', dpi=150)
plt.show()
```
