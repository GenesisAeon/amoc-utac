```python
import matplotlib.pyplot as plt
import numpy as np
from amoc_utac.system import AmocUTAC

system = AmocUTAC(seed=42)
result = system.run_cycle(duration_years=120)

crep = result['crep_state']
utac = result['utac_state']

print('CREP State:')
for k, v in crep.items():
    print(f'  {k} = {v:.4f}')
print(f"\nΓ_AMOC = {crep['Gamma']:.4f}  (target: 0.251)")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=False)

# AMOC time series
years = np.array(result['years'])
amoc  = np.array(result['amoc_sv'])
axes[0].plot(years, amoc, lw=1.5, color='steelblue', alpha=0.9, label='AMOC [Sv]')
axes[0].axhline(9.0, color='tomato', ls='--', lw=1.5, label='50% weakening (9 Sv)')
axes[0].set_ylabel('AMOC strength [Sv]')
axes[0].set_title('RAPID-calibrated synthetic AMOC time series')
axes[0].legend()

# UTAC trajectory
traj = result['utac_trajectory']
t_utac = np.array(traj['t']) + 2024
axes[1].plot(t_utac, traj['H'],      lw=2, color='steelblue', label='H(t) — UTAC ODE')
axes[1].plot(t_utac, traj['H_star'], lw=1.5, ls='--', color='darkorange', label='H*(t) — tipping setpoint')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('AMOC strength [Sv]')
axes[1].set_title('UTAC ODE trajectory (SSP2-4.5 scenario)')
axes[1].legend()

plt.tight_layout()
plt.savefig('amoc_utac_overview.png', dpi=150)
plt.show()
```
